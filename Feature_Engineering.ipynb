### Importing Libraries

```python
import pandas as pd
import statistics
import scipy.stats as stats
import numpy as np
from itertools import permutations
from tqdm import tqdm
import operator
```

###  Importing Data

```python
df = pd.read_csv('dont-overfit-ii/train.csv')
df.head()
```

### Feature Engineering using arthmetic of all feature combination and trignometric operation

```python
df_fe = df.copy()

for i in tqdm(list(permutations(df.columns.values[2:],2))):
    #https://stackoverflow.com/questions/34312846/iterate-through-a-sequence-of-operators
    for op in [('+', operator.add), ('-', operator.sub), ('*', operator.mul), ('/', operator.truediv)]:
        if((i[1]+op[0]+i[0] in df_fe.columns) and (op[0] == '+' or op[0] == '*')):
            continue
        df_fe[i[0]+op[0]+i[1]] = op[1](df_fe[i[0]],df_fe[i[1]]) #applying arithmetic operation
```

```python
a = [np.cos, np.sin, np.cosh, np.sinh, np.exp]
for i in tqdm(a):
    prefix = str(i).split('\'')[1]+'_' #prefix will be "methodName_" Eg: cos_0
    for col in df.columns[2:]: # columns from 0 to 299
        df_fe[prefix+col] = df[col].apply(i) #applying trignometric operation
```

### Removing rows with inf, -inf, nan

```python
indices_to_keep = ~df_fe.isin([np.nan, np.inf, -np.inf]).any(1)
indices_to_keep
```

```python
df_fe_clean = df_fe[indices_to_keep].astype(np.float64)
```
